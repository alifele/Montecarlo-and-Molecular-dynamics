```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML

```

```python
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim(( 0,L))
ax.set_ylim((0,L))
N=100
L = 10
pos = np.random.random((N,2))*L
vel = np.random.random((N,2))*2 - 1
data = np.hstack((pos,vel))
line, = ax.plot([],[],'8', lw=2)
```

```python
def position_update():
    data[:,0] += 0.2*data[:,2]
    data[:,1] += 0.2*data[:,3]
```

```python
def distance(data):
  mat = np.zeros((N,N))
  for i in range(N):
    for j in range(i,N):
      mat[i,j] = ((data[:,0][i] - data[:,0][j])**2 + (data[:,1][i] - data[:,1][j])**2)**(0.5)
      mat[j,i] = mat[i,j]
  return mat
```

```python
from scipy.spatial.distance import pdist, squareform
```

```python
a = np.array([[1,2,3,4,5],[3,1,2,3,1]])
squareform(pdist(a))
```

```python
def hit():
    mat = distance(data)
    for i in range(N):
      data[:,2][(mat[i]<0.1) * (mat[i] !=0)]*=-1
      data[:,3][(mat[i]<0.1) * (mat[i] !=0)]*=-1
```

```python
def wallbounce():
    data[:,2][data[:,0]>= L] *= -1
    data[:,2][data[:,0]<= 0] *= -1
    data[:,3][data[:,1]>=L] *= -1
    data[:,3][data[:,1]<=0] *= -1
```

```python
# initialization function: plot the background of each frame
def init():
    line.set_data([],[])
    return (line,)
```

```python
# animation function. This is called sequentially

def animate(i):
   
    #x = np.linspace(0, 2, 1000)
    #y = np.sin(2 * np.pi * (x - 0.01 * i))
    position_update()
    wallbounce()
    hit()
    line.set_data(data[:,0] ,data[:,1])
    return (line,)
```

```python
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=100, blit=True)
```

```python
HTML(anim.to_html5_video())
```
