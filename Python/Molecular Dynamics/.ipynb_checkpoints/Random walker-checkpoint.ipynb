```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
%matplotlib inline
```

```python
fig, ax = plt.subplots()
ax.set_xlim((-100,100))
ax.set_ylim((-100,100))
line, = ax.plot([0],[0],lw=1)
pos = [(0,0)]
```

```python
def init():
    line.set_data([0],[0])
    return line,
```

```python
def animate(i):
    rand = np.random.randint(2,size=2)*2-1
    pp = np.array(pos[-1])+rand
    pos.append(pp)
    line.set_data(np.array(pos)[:,0], np.array(pos)[:,1])
    return line,
```

```python
anim = animation.FuncAnimation(fig, animate, init_func = init,
                              frames=1000, interval=100, blit=True)
```

```python
HTML(anim.to_html5_video())
```
