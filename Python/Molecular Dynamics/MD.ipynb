```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
N = 20
L = 10
data = np.zeros((N,4))
```

```python
data[:,0] = np.random.random(N)*L
data[:,1] = np.random.random(N)*L
data[:,2] = np.random.random(N)-0.5
data[:,3] = np.random.random(N)-0.5
```

```python
plt.plot(data[:,0],data[:,1],'8')
```

```python
fig, ax = plt.subplots()
ax.set_xlim((-0.5,L+0.5))
ax.set_ylim((-0.5,L+0.5))
line, = ax.plot(data[:,0],data[:,1],'o')
```

```python
def wall_hit():
    data[:,2][np.array(data[:,0]>L ) + np.array(data[:,0]<0)] *=-1
    data[:,3][np.array(data[:,1]>L ) + np.array(data[:,1]<0)] *=-1
```

```python
def move():
    data[:,0]+=data[:,2]
    data[:,1]+=data[:,3]
```

```python
def init():
    line.set_data(data[:,0],data[:,1])
    return line,
```

```python
def animate(i):
    move()
    wall_hit()
    line.set_data(data[:,0],data[:,1])
    return line,
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init,
                              frames=200, interval=50, blit=True)
```
