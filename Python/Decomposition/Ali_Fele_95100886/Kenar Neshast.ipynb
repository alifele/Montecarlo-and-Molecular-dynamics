```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def surface_generator(surface,L,t):  
    place_list = [i for i in range(L)]
    for i in range(t):
        
        rand = int(np.random.choice(place_list,1))

        
        if rand == 0:
            if surface[rand+1][-1][1]>surface[rand][-1][1]:
                surface[rand].append((rand, surface[rand-1][-1][1]))
            elif surface[L-1][-1][1]>surface[rand][-1][1]:
                surface[rand].append((rand, surface[L-1][-1][1]))
            else:
                surface[rand].append((rand, surface[rand][-1][1]+1))
                
        elif rand == L-1:
            if surface[rand-1][-1][1] > surface[rand][-1][1]:
                surface[rand].append((rand, surface[rand-1][-1][1]))
            elif surface[0][-1][1] > surface[rand][-1][1]:
                surface[rand].append((rand, surface[0][-1][1]))
            else:
                surface[rand].append((rand, surface[rand][-1][1]+1))

        else:
            
            
            if surface[rand+1][-1][1] > surface[rand][-1][1]:
                surface[rand].append((rand, surface[rand+1][-1][1]))
                
            
            elif surface[rand-1][-1][1]> surface[rand][-1][1]:
                surface[rand].append((rand, surface[rand-1][-1][1]))
                
            else:
                surface[rand].append((rand, surface[rand][-1][1]+1))
    plot_points = []
    for place in surface:
        plot_points+=place
    return np.array(plot_points), surface
```

```python
def surface_points_extractors(data):
    height_of_surface  =[]
    for i in range(L):
        height_of_surface.append(data[data[:,0]==i][-1][1])
    return np.array(height_of_surface)
```
