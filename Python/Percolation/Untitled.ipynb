```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
y = x + 0.6*np.random.random((x.shape))
plt.plot(x,y,'8')
```

```python
from scipy import optimize
def test_func(x,a,b):
    return a*x+b

params, params_covariance = optimize.curve_fit(test_func, x, y,
                                              p0=[2,2])

print(params)
```

```python
plt.plot(x, test_func(x,params[0], params[1]))
plt.plot(x,y,'8')
```
