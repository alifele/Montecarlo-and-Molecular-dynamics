# Simple Sampeling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,5,0.01)
yy = x**2
plt.plot(x,yy,label='$y=x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
def y(x):
    return x**2
```

```python
rand_list_upper = []
rand_list_lower = []
for n in range(500):
    i= np.random.random()*5
    j= np.random.random()*25
    if j < y(i):
        rand_list_lower.append((i,j))
    else :
        rand_list_upper.append((i,j))
rand_list_upper = np.array(rand_list_upper)
rand_list_lower = np.array(rand_list_lower)
plt.plot(x,yy,label='$y=x^2$',linewidth=4)
plt.plot(rand_list_upper[:,0],rand_list_upper[:,1],'8g',alpha=0.6,label='The simple samples')
plt.plot(rand_list_lower[:,0],rand_list_lower[:,1],'8b',alpha=0.6,label='The simple samples')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

```python
counter=0
N =1000
integral = []
for many_values in range(50):
    counter=0
    for i in range(N):
        rand_x = np.random.random()*5
        rand_y = np.random.random()*25
        if rand_y<y(rand_x):
              counter+=1
    integral.append(counter/1000 * 25*5)
```

```python
print('the result of integral is :{}'.format(counter/1000 * 25*5))
print("The Real Value of the integral is :{}".format(125/3))
```

```python
mean = np.array(integral).mean()
```

```python
plt.plot(integral)
plt.plot(integral,'8',label='Values of integral')
plt.plot([mean for i in range(len(integral))],label='Mean = {}'.format(mean))
plt.plot([41.66 for i in range(len(integral))],label='Real Value=%3.4f'%(125/3))
plt.xlabel('Trials')
plt.ylabel('Value of Integral')
plt.legend()
```

```python
CM_values = []
for many_values in range(50):
    Total_mass = 0
    CM=0  
    for sample in range(10000):
        r_rand = np.random.random()*2 - 1
        th_rand = np.random.random() *np.pi
        z = r_rand*np.cos(th_rand)
        dm = dv*density(z)
        Total_mass += dm
        CM += r_rand*np.cos(th_rand)*dm
    CM_values.append(CM/Total_mass)
```

```python
%%time 
# Simple Samples
N =25600
Guassian_0 = Guassian(0,1,0)
integral = []
for many_values in range(10):
    counter  = 0
    for sample in range(N):
        rand_x = np.random.random()*2
        counter+=Guassian(rand_x,1,0)/Guassian_0
    integral.append(counter/N*2*Guassian_0)
    
mean = np.array(integral).mean()
std = np.array(integral).std()
print('N = {}'.format(N))
print('mean = {}'.format(mean))
print('error = {}'.format(std))
print('delta = {}'.format(mean-0.477249868))
```
