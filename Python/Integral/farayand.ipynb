```python
def cauchi(x,mean,gama):
    return gama/((x-mean)**2 + gama**2 )/ np.pi
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.01)
for gamma in range(1,10):
    y = cauchi(x,0,gamma)
    plt.plot(x,y)
```

```python
x = np.linspace(-100,100,10000)
y = cauchi(x,5,4)
plt.plot(x,y)
```
