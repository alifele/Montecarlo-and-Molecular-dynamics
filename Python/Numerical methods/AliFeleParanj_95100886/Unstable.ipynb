```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
func = []
Q = 0
Qp = 1/3 * (10-Q/1)
t=0
V = 10
h = 0.1
R = 3
t_stop = 20
func.append((t,Q))
Q = Q + h*(V-Q)/R
t +=h
Qp = 1/3 * (10-Q/1)
func.append((t,Q))
    
```

```python
for i in range(int(t_stop/h)):
    Q = func[-2][1] + 2*Qp*h
    Qp = 1/3 * (10-Q/1)
    t+=h
    func.append((t,Q))
```

```python
data = np.array(func)
plt.plot(data[:,0],data[:,1])
plt.title('the unstabality of the algorithm\nCharging Capacitor',size=15)
plt.xlabel('time[$10^{-3}C $]',size=15)
plt.ylabel('Charge[$10^{-6}C $]',size=15)
```

$y_{n+1} = y_{n-1} + 2 y'_n h$


# Discharging

```python
for i in range(int(t_stop/h)):
    Q = func[-2][1] + 2*Qp*h
    Qp = 1/3 * (10-Q/1)
    t+=h
    func.append((t,Q))
```

```python
data = np.array(func)
plt.plot(data[:,0],data[:,1])
plt.title('the unstabality of the algorithm\nDischarging Capacitor',size=15)
plt.xlabel('time[$10^{-3}C $]',size=15)
plt.ylabel('Charge[$10^{-6}C $]',size=15)
```
