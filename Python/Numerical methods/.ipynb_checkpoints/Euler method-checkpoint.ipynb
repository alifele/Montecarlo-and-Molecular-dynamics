```python
import numpy as np
import matplotlib.pyplot as plt
```

the units are:

[R] = 1000 ohm

[C] = 10^-6 C/V

[V] = 1 V

[Q] = 10^-6 C

[t] = 1 ms

(the units are related to each other by C=Q/V and V=IR=dQ/dt * R )

```python
R = 3
V = 10
Q_0 = 0 
t_0 = 0
Q = Q_0
t = t_0
h = float(0.00001)
t_final = 0.5
func = []
func.append((t_0,Q_0))
for i in range(int(t_final/h)):
    Q = Q + h*(V-Q)/R
    t +=h
    func.append((t,Q))
    
```

```python
data = np.array(func)
plt.title('Capacity charging\nwith h = {}'.format(h))
plt.plot(data[:,0],data[:,1],label = "the Euler's method reults")
plt.plot(data[-1,0],data[-1,1],'o',markersize=8,label="the charge of the capacitor after 5 ms\nQ=%3.5f$*10^{-6}$"%(data[-1,1]))
plt.xlabel('time[$10^{-3}$s]',size=13)
plt.ylabel('Charge[$10^{-6}C$]',size=13)
plt.legend(prop={'size':10.8},loc=(0.01,0.75))
```

we know that the exact solution is :

$Q = CV(1-e^{-t/RC})$

```python
def F(t):
    return 10*(1-np.exp(-t/3))
```

```python
plt.plot(data[::100,0],data[::100,1],'o')
plt.plot(data[:,0],F(data[:,0]))
```

# Calcualting the Error Vs h

```python
R = 3
Error = 0
V = 10
Q_0 = 0 
t_0 = 0
Q = Q_0
t = t_0
t_final = 0.5
E = []
H = [float(0.91**i) for i in range(1,18)]
for h in H:
    for i in range(int(t_final/h)):
        Q = Q + h*(V-Q)/R
        t +=h
    Error = np.abs(Q - F(0.5))
    E.append((h,Error))
data = np.array(E)
plt.semilogx(data[:,0],data[:,1,],'o')
plt.semilogx(data[:,0],data[:,1,])
plt.title('the SemiLogx plot of Error vs. H',size=17)
plt.xlabel('h',size=15)
plt.ylabel('Error',size=15)
plt.grid('ON')
    
```
