```python
import numpy as np
import matplotlib.pyplot as plt
```

the units are:

[R] = 1000 ohm

[C] = 10^-6 C/V

[V] = 1 V

[Q] = 10^-6 C

[t] = 1 ms

(the units are related to each other by C=Q/V and V=IR=dQ/dt * R )

```python
R = 3
V = 1
Q_0 = 0 
t_0 = 0
Q = Q_0
t = t_0
h = 0.01
t_final = 0.5
func = []
func.append((t_0,Q_0))
for i in range(int(t_final/h)):
    Q = Q + h*(1-Q)/R
    t +=h
    func.append((t,Q))
    
```

```python
data = np.array(func)
plt.title('Capacity charging\nwith h = {}'.format(h))
plt.plot(data[:,0],data[:,1],label = "the Euler's method reults")
plt.plot(data[-1,0],data[-1,1],'o',markersize=8,label="the charge of the capacitor after 5 ms\nQ=%3.5f$*10^{-6}$"%(data[-1,1]))
plt.xlabel('time[$10^{-3}$s]',size=13)
plt.ylabel('Charge[$10^{-6}C$]',size=13)
plt.legend(prop={'size':10.8},loc=(0.01,0.75))
```
