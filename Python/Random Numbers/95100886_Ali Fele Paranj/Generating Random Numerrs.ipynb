# Evaluating the Mean of Random Generator

```python
data = []
samples = [1.7**i for i in range(25)]
Q=50
for sample_size in samples :
    mean = 0
    std = 0
    for q in range(Q):
        a = np.random.random(int(sample_size))
        mean += a.mean()
        std += a.std()
    data.append((sample_size, mean/Q, std/Q))
```

```python
data = np.array(data)
```

```python
plt.loglog(data[:,0],data[:,1])
plt.loglog(data[:,0],data[:,1],'8')
plt.xlabel('samples')
plt.ylabel('mean')
```

```python
plt.loglog(data[:,0],data[:,2])
plt.loglog(data[:,0],data[:,2],'8')
plt.xlabel('samples')
plt.ylabel('std')
```

```python
cor_data= []
Q = 20
tau_list= [i for i in range(1,120)]
for tau in tau_list :
    Cor=0
    for q in range(Q):
        cor=0
        rand_data = np.random.random(5000)
        
        for i in range(rand_data.shape[0]-tau):
            cor += rand_data[i]*rand_data[i+tau]
            
        Cor += (cor/(5000-tau) - rand_data.mean()**2)/rand_data.std()**2
        #Cor += (cor/(5000-tau))
    cor_data.append((tau,Cor/Q))
    
cor_data = np.array(cor_data)
```

```python
plt.figure(figsize=(10,5))
plt.plot(cor_data[:,0], cor_data[:,1])
plt.plot([cor_data[:,1].mean() for i in range(cor_data[:,1].shape[0])], label='Mean')
plt.legend()
plt.plot(cor_data[:,0], cor_data[:,1],'o')
plt.xlabel('$\\tau$')
plt.ylabel('$R(\\tau)$')
```

# Generating the Guassian Random Nmubers

```python
sigma = 10
guassian_random = []
for i in range(10000):
    th = 2*np.pi*np.random.random()
    r = (-2*sigma**2*np.log(1-np.random.random()))**0.5
    guassian_random.append(r*np.sin(th))
```

```python
plt.figure(figsize=(14,10))
n, bins, patch= plt.hist(guassian_random,bins=30,label='Guassian with mean=0 \nand $\\sigma={}$'.format(sigma))
plt.title('Generating the Guassian numbers')
plt.legend(prop={'size': 20})
```

```python
bins2 = np.zeros(n.shape)
for i in range(int(n.shape[0])):
    bins2[i] = (bins[i]+ bins[i+1])/2
bins = bins2
```

```python
def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) / 4 / stddev)**2)
```

```python
parameters,_ = optimize.curve_fit(gaussian,bins,n)
```

```python
plt.figure(figsize=(12,8))
plt.plot(bins,gaussian(bins,*parameters),label='Guassina fit with $\\sigma$=%3.4f'%parameters[-1]**2,linewidth=4)
n, bins, patch= plt.hist(guassian_random,bins=30,label='Guassian $\\sigma={}$'.format(sigma))
plt.legend(prop={'size':15})
```

# Centeral Limit theorem

```python
plt.figure(figsize=(14,10))
Q = 1000
N = 7000
counter=0
Qlist = [2**i for i in range(4)]
G_data=[]
for Qsize in Qlist:
    Q = Qsize
    G_data = []
    for i in range(N):
        counter =0
        for j in range(Q):
            counter+=np.random.random()
        G_data.append(counter/Q)
    n,bins, patch = plt.hist(G_data,bins=30,label='the size of the minibatches={}'.format(Q))
plt.legend()
```

```python
N = 300
Q_list = [100*i for i in range(1,10)]
var_data = []
for Q in Q_list:
    var_1 = 0 
    data = []
    for n in range(N):
        rand = np.random.random(Q)
        var_1 += rand.var()/N
        data.append(rand.mean())
        
    var_data.append((Q,np.array(data).var(),var_1))
  
```

```python
data = np.array(var_data)
plt.plot(data[:,0],data[:,1])
plt.plot(data[:,0],data[:,1],'8')
plt.xlabel('N')
plt.ylabel('Var of Guassian')
```

```python
from scipy import optimize
```

```python
def sigma(x,var):
    return var/x
```

```python
par,_ = optimize.curve_fit(sigma,data[:,0],data[:,1])
par
```

```python
plt.figure(figsize=(14,10))
plt.plot(data[:,0], sigma(data[:,0],*par),label='$\\sigma^2$ = %3.4f/N\n(fit)'%(par[0]),linewidth=5)
plt.plot(data[:,0],sigma(data[:,0],data[:,2]),'s',label='uniform_rand_var/N')
plt.plot(data[:,0],data[:,1],'8',label='real data')
plt.legend(prop={'size': 20})
plt.xlabel('N',size=20)
plt.ylabel('$\\sigma^2$',size=20)

```

```python
plt.figure(figsize=(14,10))
Q = 1000
N = 7000
counter=0
Qlist = [2**i for i in range(4)]
G_data=[]
for Qsize in Qlist:
    Q = Qsize
    G_data = []
    for i in range(N):
        counter =0
        for j in range(Q):
            counter+=np.random.triangular(-3, 0, 8)
        G_data.append(counter/Q)
    n,bins, patch = plt.hist(G_data,bins=100,label='the size of the minibatches={}'.format(Q))
plt.legend()
plt.title('Ceneral limit theorem with triangular distribution')
```
