```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Generator(t,L): # t is the number of particles released and L is the dimension of the lattice
  surface = np.zeros((L,))
  for i in range(t):
    rand = int(np.random.random()*L)
    surface[rand] +=1
    
  return surface
    
```

```python
def display(subsurface_points, surface, marker):
  points = []
  for i in range(surface.shape[0]):
    
    for j in range(int(surface[i])):
      points.append((i,j + subsurface_points[subsurface_points[:,0]==i][-1][1] ))
    
  points = np.array(points)
  plt.plot(points[:,0], points[:,1], marker)
  return points
  
```

```python
L = 50
t = 1000
subsurface_points = np.array([(i,0) for i in range(L)])
surface = Generator(t, L)
points = display(subsurface_points,surface,'bs')

surface=Generator(t, L)
subsurface_points = points 
points = display(subsurface_points, surface, 'rs')


surface=Generator(t, L)
subsurface_points = points 
points = display(subsurface_points, surface, 'gs')


```

```python
def Generator(subsurface,t,L): # t is the number of particles released and L is the dimension of the lattice
  surface = subsurface
  for i in range(t):
    rand = int(np.random.random()*L)
    surface[rand] +=1
    
  return surface
    
```

```python
L = 100
subsurface = np.zeros((L,))

surface = Generator(subsurface, 1000, L)
```

```python
t_list = [2**i for i in range(1,20)]
subsurface = np.zeros((L,))
std_list = []
N_total=0
for t in t_list:
    N_total +=t
    surface = subsurface
    surface = Generator(subsurface, t, L)
    std_list.append((N_total,surface.std()))
```

```python
from sklearn.linear_model import LinearRegression
std_list = np.array(std_list)
loglog_std_list = np.log(std_list)
model = LinearRegression()
Features = loglog_std_list[:,0].reshape(-1,1)
y = loglog_std_list[:,1]
model.fit(Features,y)
beta = model.coef_
bias = model.intercept_
```

```python
std_list = np.array(std_list)
plt.plot(loglog_std_list[:,0], loglog_std_list[:,1],'o')
plt.plot(beta*loglog_std_list[:,0]+bias, label = '{}*t+{}'.format(beta, bias))
#plt.loglog(std_list[:,0], std_list[:,1], linewidth=3)

plt.xlabel('Number of particles')
plt.ylabel('w (standard deviation)')
plt.legend()
```
