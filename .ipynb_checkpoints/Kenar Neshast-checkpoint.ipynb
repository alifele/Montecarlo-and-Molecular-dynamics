```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def surface_generator(surface,L,t):  
  
    for i in range(t):
        
        rand = int(np.random.random()*L)

        if rand == L-1:
            if surface[rand-1][-1][1] > surface[rand][-1][1]:
                surface[rand].append((rand, surface[rand-1][-1][1]))
            elif surface[0][-1][1] > surface[rand][-1][1]:
                surface[rand].append((rand, surface[0][-1][1]))
            else:
                surface[rand].append((rand, surface[rand][-1][1]+1))

        elif rand == 0:
            if surface[rand+1][-1][1]>surface[rand][-1][1]:
                surface[rand].append((rand, surface[rand-1][-1][1]))
            elif surface[L-1][-1][1]>surface[rand][-1][1]:
                surface[rand].append((rand, surface[L-1][-1][1]))
            else:
                surface[rand].append((rand, surface[rand][-1][1]+1))

        else:
            if surface[rand+1][-1][1] > surface[rand][-1][1]:
                surface[rand].append((rand, surface[rand+1][-1][1]))
            elif surface[rand-1][-1][1]> surface[rand][-1][1]:
                surface[rand].append((rand, surface[rand-1][-1][1]))
            else:
                surface[rand].append((rand, surface[rand][-1][1]+1))
    plot_points = []
    for place in surface:
        plot_points+=place
    return np.array(plot_points)
```

```python
L=200
surface = [ [] for i in range(L)]
i=0
for place in surface:
    place.append((i,0))
    i+=1
data = surface_generator(surface,L,15000)
```

```python
plt.figure(figsize=(10,10))
plt.plot(data[:,0], data[:,1],'s')
```

```python
L=200
surface = [ [] for i in range(L)]
i=0
for place in surface:
    place.append((i,0))
    i+=1
surface[int(L/2)].append((50,L))
```

```python
data = surface_generator(surface,L,15000)
```

```python
plt.figure(figsize=(10,10))
plt.plot(data[:,0], data[:,1],'s', markersize=2)
```
