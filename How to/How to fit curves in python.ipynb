```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

# Guassian Fit

```python
a = np.random.randn(100,100).reshape(-1,)
a = a + 10*np.random.random(np.shape(a))
```

```python
n, bins,pa = plt.hist(a, bins=30)
```

```python
plt.plot(bins[:-1],n,'o')
```

```python
def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) / 4 / stddev)**2)
```

```python
parameters,_ = optimize.curve_fit(gaussian,bins[:-1],n)
```

```python
parameters
```

```python
plt.plot(bins[:-1],gaussian(bins[:-1],*parameters),label='Gaussian(mu=%3.2f,std=%3.2f)'%(parameters[0],parameters[1]))
plt.plot(bins[:-1],n,'or')
plt.legend()
```

# Sine Fit

```python
x = np.linspace(0,10,100)
y = np.sin(x) + np.random.random(x.shape)
```

```python
plt.plot(x,y)
```

```python
def sin(x,A,w,phi,bias):
    return A*np.sin(A*x+phi)+bias
```

```python
parameters,_ = optimize.curve_fit(sin,x,y)
```

```python
parameters
```

```python
plt.plot(x,sin(x,*parameters))
plt.plot(x,y)
```
