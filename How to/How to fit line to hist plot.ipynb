# How to plot line fitted to the hist plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rand_data = []
for i in range(1000):
    rand_data.append(np.random.randn())
```

```python
n, bins , patch = plt.hist(rand_data)
```

```python
bins2 = np.zeros(n.shape)
for i in range(int(n.shape[0])):
    bins2[i] = (bins[i]+ bins[i+1])/2
bins = bins2
```

```python
plt.hist(rand_data)
plt.plot(bins, n)
```
