```python
Lx, Ly =100,100
```

```python
J = 1 * Jcrit
H = 0
```

```python
sigma = np.ones((Lx,Ly),dtype = 'i4')
```

```python
plt.figure()
im = plt.imshow(sigma, vmin=-1, vmax=1)
```

```python
def energy(lattice,J):
    energy = 0 
    L = lattice.shape[0]
    for i in range(lattice.shape[0]):
        for j in range(lattice.shape[1]):
            energy+= -J*lattice[i,j]*(lattice[(i-1)%L,j]+lattice[(i+1)%L,j]+lattice[i,(j-1)%L]+lattice[i,(j+1)%L])

    return energy/2
```

# filip

```python
lattice = np.ones((20,20))
L = lattice.shape[0]
```

```python
 def flip(lattice, T):
    L = lattice.shape[0]
    T = T/L**2
    for n in range(1000):
        i  = np.random.randint(0,L)
        j = np.random.randint(0,L)
        E1 = energy(lattice,1)
        lattice[i,j] = -lattice[i,j]
        E2 = energy(lattice,1)
        dE = E2-E1
        if dE < 0 :
            continue
        if dE > 0:
            rand = np.random.random()
            if rand < np.exp(-dE/T):
                continue
            else:
                lattice[i,j] = -lattice[i,j]
    return lattice
```

```python
flip(lattice,1000)
```

```python
plt.imshow(lattice)
```

```python
T = [1.2**i for i in range(16)]
for t in T:
    fig  = plt.figure(figsize=(15,15))
    for i in range(16):
        ax = fig.add_subplot(4,4,i+1)
        ax.imshow(flip(lattice,t))
```
