```python
import matplotlib.pyplot as plt
import numpy as np


plt.ion()
x=0
y=0
data=[]
L = 10000
for i in range(L):
    rand = np.random.random()
    
    if rand>0 and rand<0.25:
        x = x+1
    if rand>0.25 and rand<0.5:
        x = x-1
    if rand>0.5 and rand<0.75:
        y = y+1
    if rand>0.75 and rand<1:
        y = y-1
    x = x+0.1*np.random.randn()
    y = y+0.1*np.random.randn()
    data.append((x,y))
    
data = np.array(data)
plt.figure(figsize=(8,8))
plt.axis('equal')
plt.axis('off')
plt.plot(data[:,0], data[:,1], label='random walker with {} steps'.format(L))

plt.legend()


```

# Probability plots of random walker

```python
def run_random_walker(N,Q):
    end_points = []
    for ensamble in range(Q):
        x=0
        y=0
        for walk in range(N):
            rand = np.random.random()
            if rand>0 and rand<0.25:
                x = x+1
            if rand>0.25 and rand<0.5:
                x = x-1
            if rand>0.5 and rand<0.75:
                y = y+1
            if rand>0.75 and rand<1:
                y = y-1
        x = x+0.1*np.random.randn()
        y = y+0.1*np.random.randn()
        end_points.append((x,y))

    return np.array(end_points)
    
```

```python
def run_random_walker1_D(N,Q):
    end_points = []
    for ensamble in range(Q):
        x=0
        y=0
        for walk in range(N):
            rand = np.random.random()
            if rand>0 and rand<0.54:
                x = x+1
            if rand>0.54 and rand<1:
                x = x-1
           # if rand>0.5 and rand<0.75:
            #    y = y+1
            #if rand>0.75 and rand<1:
             #   y = y-1
        x = x+0.1*np.random.randn()
        #y = y+0.1*np.random.randn()
        end_points.append((x,y))

    return np.array(end_points)
    
```

```python
Q = 1000
N = 100
data = run_random_walker(N,Q)
```

# Dynamic of sigma

```python
Q=50
N_list = [10*i for i in range(1,50)]
std_dynamic=[]
mean_dynamic=[]
for N in N_list:
    data = run_random_walker(N,Q)
    n, bins, pa = plt.hist(data[:,0],bins=30)
    popt, _ = optimize.curve_fit(gaussian, bins[:-1], n)
    mean = popt[1]
    std = popt[2]
    std_dynamic.append((N,std))
    mean_dynamic.append((N,mean))
```

```python
Q=500
N_list = [10*i for i in range(1,50)]
std_dynamic=[]
mean_dynamic=[]
for N in N_list:
    data = run_random_walker1_D(N,Q)
    n, bins, pa = plt.hist(data[:,0],bins=30)
    popt, _ = optimize.curve_fit(gaussian, bins[:-1], n)
    mean = popt[1]
    std = popt[2]
    std_dynamic.append((N,std))
    mean_dynamic.append((N,mean))
```

```python
std_dynamic = np.array(std_dynamic)
mean_dynamic = np.array(mean_dynamic)
```

```python
plt.plot(std_dynamic[:,0], np.abs(std_dynamic[:,1])**2,'8')
plt.xlabel('N')
plt.ylabel('Variance')
```

```python
def linear_fit(x,a,b):
    return a*x + b
```

```python
popt, _ = optimize.curve_fit(linear_fit,std_dynamic[:,0], np.abs(std_dynamic[:,1])**2)
```

```python
plt.plot(std_dynamic[:,0], np.abs(std_dynamic[:,1])**2,'8')
plt.xlabel('N')
plt.ylabel('Variance')
plt.plot(std_dynamic[:,0],linear_fit(std_dynamic[:,0],*popt), label='%3.2f*x+%3.2f\np=%3.2f,q=%3.2f'%(popt[0],popt[1],0.54,0.46))
plt.legend()
plt.title('Dynamic of variance')
```

```python
popt, _ = optimize.curve_fit(linear_fit,mean_dynamic[:,0], mean_dynamic[:,1])
plt.plot(mean_dynamic[:,0], mean_dynamic[:,1]+1,'8')
plt.xlabel('N')
plt.ylabel('Mean')
plt.plot(mean_dynamic[:,0],linear_fit(mean_dynamic[:,0],*popt)+1, label='%3.2f*x+%3.2f\np=%3.2f,q=%3.2f'%(popt[0],popt[1],0.54,0.46))
plt.legend()
plt.title('Dynamic of mean')
```

# Random Walker as Time series

```python
Walkers = []
for ensamble in range(3000):
    pos = 0
    data = []
    for t in  range(1000):
        rand = np.random.random()
        if rand<0.5:
            pos+=1
        if rand>0.5:
            pos-=1

        data.append((t,pos))
    Walkers.append(np.array(data))
```

# Random Walker with Trap

```python
max_data = []
mean_data = []
Trap = [i for i in range(2,30)]
for trap in Trap:
    data= []
    max_life_time = 0
    avr_life_time = 0
    for trap_ensamble in range(30):
        data=[]
        for Walker in range(1000):
            pos = 0
            for t in range(1000):
                rand = np.random.random()
                if rand<0.5:
                    pos+=1
                if rand>0.5:
                    pos-=1
                if pos >trap or pos < -trap:
                    data.append((Walker,t))
                    break
                    
        data = np.array(data)    
        n,bins,patch = plt.hist(data[:,1],bins=50)
        max_life_time += bins[n.argmax()]
        avr_life_time += data[:,1].mean()
    max_data.append((trap,max_life_time/30))
    mean_data.append((trap,avr_life_time/30))
```

```python
max_data = np.array(max_data)
mean_data = np.array(mean_data)
```

```python
plt.plot(max_data[:,0],max_data[:,1],'8')
plt.xlabel('bar size')
plt.ylabel('max of life time')
```
