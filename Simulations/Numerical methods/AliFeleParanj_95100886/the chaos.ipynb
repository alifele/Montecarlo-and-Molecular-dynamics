```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
r = np.arange(0,1,0.00001)
for R in r:
    x1 = np.random.random()
    for run in range(700):
        x1 = 4*R*x1*(1-x1)
    for sample in range(50):
        x1 = 4*R*x1*(1-x1)
        x.append((R,x1))
    
```

```python
X = np.array(x)
plt.figure(figsize=(54,18))
plt.plot(X[:,0],X[:,1],'o',markersize=1.5)
plt.title('the Bifurcation Graph',size = 45)
plt.xlabel('r',size=40)
plt.ylabel('$X_n$',size=40)
```

```python
for i,elem in enumerate(X[1347300:]):
    if elem[1] - X[i+1347300+4][1]>0.40:
        print(i)
        break
```
