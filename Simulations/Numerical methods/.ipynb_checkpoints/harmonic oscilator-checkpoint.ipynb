```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_0 = 0
x_1 = 2
t = 0
t_stop = 60
x = []
v_list = []
x.append((t,x_0))
h = 0.5
```

## the Euler method

```python
for i in range(int(t_stop/h)):
    x_0 = x_0 + h*x_1
    v_list.append((x_0,x_1))
    x_1 = x_1 -h*x_0
    t+=h
    x.append((t,x_0))
    
```

```python
data = np.array(x)
plt.figure(figsize=(15,5))
plt.plot(data[::30,0],data[::30,1],'o',label='the simulation values')
plt.plot(data[:,0],2*np.sin(data[:,0]),linewidth=3,alpha=0.7,label='the actual values')
plt.ylabel('position')
plt.xlabel('time')
plt.title('The results of the simulation with Euler method\nh={}'.format(h))
plt.legend(loc=3)

```

```python
v_data = np.array(v_list)
plt.plot(v_data[:,0],v_data[:,1],'.')
plt.axis('equal')
plt.xlabel('position')
plt.ylabel('velocity')
plt.title('the v-x plot of the results of the Eurler method\n with the h={}'.format(h))
```

## Euler Cromer method

$v[n+1] = v[n] + a[n]*dt$

$x[n+1] = x[n] + v[n+1]*dt$

```python
v = 2
x = 0
a = -x
dt = 0.5
t_stop = 60
t = 0
x_list = []
v_list = []
x_list.append((t,x))
```

```python
for i in range(int(t_stop/dt)):
    a = -x
    v += a*dt
    x += v*dt
    v_list.append((x,v))
    t+=dt
    x_list.append((t,x))
```

```python
data = np.array(x_list)
plt.figure(figsize=(15,5))
plt.plot(data[::30,0],data[::30,1],'o',label='the simulation values')
plt.plot(data[:,0],2*np.sin(data[:,0]),linewidth=3,alpha=0.7,label='the actual values')
plt.ylabel('position')
plt.xlabel('time')
plt.title('The results of the simulation with Euler Cromer method\nh={}'.format(dt))
plt.legend(loc=3)

```

```python
v_data = np.array(v_list)
plt.plot(v_data[:,0],v_data[:,1])
plt.axis('equal')
plt.xlabel('position')
plt.ylabel('velocity')
plt.title('the v-x plot of the results of the Eurler Cromer method\n with the h={}'.format(h))
```

## The verlet method

in order to use the verlet method we should set:

$x[1] = x[0] + v[0]dt + \dfrac{1}{2}a[0]dt^2$

and then iterate:

$x[n+1] = 2x[n] - x[n-1] + a[n]dt^2$ 

```python
v = 2
x = 0
a = -x
dt = 0.01
t_stop = 60
t = 0
x_list = []
x_list.append((t,x))
x = x + v*dt + 0.5*a*dt**2
a=-x
x_list.append((t,x))
```

```python
for i in range(int(t_stop/dt)):
    x = 2*x_list[-1][1] - x_list[-2][1] + a*dt**2
    a = -x
    t+=dt
    x_list.append((t,x))
```

```python
data = np.array(x_list)
plt.figure(figsize=(15,5))
plt.plot(data[::29,0],data[::29,1],'o',label='the simulation values')
plt.plot(data[:,0],2*np.sin(data[:,0]),linewidth=3,alpha=0.7,label='the actual values')
plt.ylabel('position')
plt.xlabel('time')
plt.title('The results of the simulation with Verlet method\nh={}'.format(dt))
plt.legend(loc=3)
```

## Velocity verlet

$x[n+1] = x[n] + v[n]dt + \dfrac{1}{2}a[n]dt^2$

$a[n+1]= -x[n+1]$ 

$v[n+1] = v[t] + \dfrac{1}{2}(a[n] + a[n+1])dt$

```python
v = 2
x = 0
a = -x
dt = 0.5
t_stop = 60
t = 0
v_list = []
x_list = []
x_list.append((t,x))
```

```python
for i in range(int(t_stop/dt)):
    x +=v*dt + 0.5*a*dt**2
    x_list.append((t,x))
    v_list.append((x,v))
    a_pre = a
    a = -x
    v += 0.5*(a_pre + a)*dt
    t+=dt
```

```python
data = np.array(x_list)
plt.figure(figsize=(15,5))
plt.plot(data[::29,0],data[::29,1],'o',label='the simulation values')
plt.plot(data[:,0],2*np.sin(data[:,0]),linewidth=3,alpha=0.7,label='the actual values')
plt.ylabel('position')
plt.xlabel('time')
plt.title('The results of the simulation with velocity Verlet method\nh={}'.format(dt))
plt.legend(loc=3)
```

```python
v_data = np.array(v_list)
plt.plot(v_data[:,0],v_data[:,1])
plt.axis('equal')
plt.xlabel('position')
plt.ylabel('velocity')
plt.title('the v-x plot of the results of the Velicity Verlet method\n with the h={}'.format(h))
```
