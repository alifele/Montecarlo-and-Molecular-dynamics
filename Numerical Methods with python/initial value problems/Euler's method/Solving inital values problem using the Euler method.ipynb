suppose that the y'(x) = 2x and the h=0.1


```python
x_0 = 2
y_0 = 5
```

```python
x_stop = 10
x = x_0
y = y_0
function = []
h=0.1
function.append((x,y))
while x<x_stop:
    y = y + h * 2*x
    x = x + h
    function.append((x,y))
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array(function)
plt.plot(data[:,0],data[:,1])
plt.show()
```

```python
x = np.arange(x_0,x_stop,h)
y = x**2 +1
plt.plot(x,y,label='the real values')
plt.plot(data[:,0],data[:,1],'.',label ='the values of the euler method')
plt.legend()
plt.title('the evaluation of the euler method with h = {}'.format(h))
```

```python
x_0 = 0
y_0 = 1
x_stop = 10
h = 0.48
x = x_0
y = y_0 
function = []
function.append((x,y))
while x<x_stop:
    y = y + (x**2-4*y)*h
    x = x+h
    function.append((x,y))
    
```

```python
data = np.array(function)
plt.plot(data[:,0],data[:,1],'o',markersize=5,label = 'the result of the euler method')
x = np.arange(x_0,x_stop,h)
y = 31/32 * np.exp(-4*x) + 0.25*x**2 - 1/8*x + 1/32
plt.plot(x,y,label = 'the real values')
plt.legend()
plt.title('the euler method with h = {}'.format(h))
```
