```python
import numpy as np
import matplotlib.pyplot as plt
```

# The first derivitive


```python
%%latex
$f'(x) = \dfrac{f(x+h)-f(x-h)}{2h} + O(h^2)$

```

```python
h = 0.01
x = np.arange(0,10,h)
y = x**2
plt.plot(x,y)
plt.title('the $x^2$ plot')
```

```python
df = [y[0]]
for i in range(1,len(y)-1):
    df.append((y[i+1]-y[i-1])/2/h)
df.append(df[-1])
```

```python
plt.plot(x,df)
plt.title('the first derivitive')
```

# The second derivitive

```python
%%latex
$f''(x) = \dfrac{f(x+h) - 2f(x) + f(x-h)}{h^2} + O(h^2)$
```

```python
h = 0.01
x = np.arange(0,3,0.01)
y = x**3
plt.plot(x,y,label='$y=x^3$')
plt.legend()
```

```python
df = []
for i in range(1,len(y)-1):
    df.append((y[i+1]-y[i-1])/2/h)
    
plt.plot(x[1:-1],df,label='the first derivitive')
plt.legend()
```

```python
d2f =[]
for i in range(1,len(y)-1):
    d2f.append(  (y[i+1]-2*y[i]+y[i-1])/h**2  )
plt.plot(x[1:-1],d2f,label='the second derivitive')
plt.legend()
```

# The third derivitive

```python
%%latex 
$f'''(x) = \dfrac{f(x+2h)-2f(x+h)+2f(x-h)-f(x-2h)}{2h^3}$
```

```python
h = 0.01
x = np.arange(0,3,h)
y = x**4
plt.plot(x,y)
```

```python
df = []
for i in range(1,len(y)-1):
    df.append((y[i+1]-y[i-1])/2/h)
    
plt.plot(x[1:-1],df,label='the first derivitive')
plt.legend()
```

```python
d2f = []
for i in range(1,len(y)-1):
    d2f.append((y[i+1]-2*y[i]+y[i-1])/h**2)
plt.plot(x[1:-1],d2f,label='the second derivitive')
```

```python
d3f = []
for i in range(2,len(y)-2):
    d3f.append( (y[i+2]-2*y[i+1]+2*y[i-1]-y[i-2])/2/h**3 )
plt.plot(x[2:-2],d3f,label = 'the third derivitive')
plt.legend()
```
